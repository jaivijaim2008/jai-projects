```python
import pandas as pd
cars_data=pd.read_csv("mtcars.csv")
```

```python
cars_data
```

## univarient analysis

```python
from matplotlib import pyplot as plt
plt.hist(cars_data["disp"])
plt.ylabel("frequency")
plt.xlabel("mpg")
plt.show()
```

## bivarient analysis

```python
import seaborn as sns
sns.barplot(data=cars_data , x="cyl",y="mpg")
plt.title("mileage per gallon")
plt.show()
```

## multivarient analysis

```python
sns.pairplot(data=cars_data)
```
