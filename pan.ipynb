## getting and knowing the data

```python
import pandas as pd
order_details=pd.read_csv(filepath_or_buffer="data.tsv.txt",sep="\t")
order_details
```

```python
order_details.shape
```

```python
order_details.isnull().sum()
```

```python
order_details.describe(include='all')
```

### all 50 items names 

```python
print(order_details['item_name'].unique())
```

```python
order_details.nunique()
```

### least 5 selling items 

```python
order_details.groupby(by="item_name")["quantity"].sum().sort_values().head(5)
```

### top 5 selling items

```python
order_details.groupby(by="item_name")["quantity"].sum().sort_values(ascending=False).head(5)
```

```python
order_details.groupby(by="item_name")["quantity"].sum()
```

### top 10 chicken bowl choice description

```python
print(order_details[order_details["item_name"] == "Chicken Bowl"]["choice_description"].value_counts().head(10))

```

### top 10 customers buying bulk orders

```python
order_details.groupby(by="order_id")["quantity"].sum().sort_values(ascending=False).head(10)
```

## top 10 highest revenue items  

```python
order_details
```

```python
order_details["item_price"]=order_details["item_price"].str.replace("$","").astype(float)
```

### top 5 revenue item prices 

```python
top_5_revenue=order_details.groupby(by="item_name")["item_price"].sum().sort_values(ascending=True).head(5)
top_5_revenue
```

### total of top 5 revenues

```python
totalrevenue=top_5_revenue.sum()
totalrevenue
```
