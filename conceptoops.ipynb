## Learnig oops concepts

```python

class employee:
    def __init__(self,name):
        self.name = name

    def login(self):
        print (f"{self.name} logged in")
    def view_tasks(self):
        print (f"{self.name} has viewed tasks")

class manager(employee):
    def assign_task(self):
        print (f"{self.name} has assigned a task")
class team_member(employee):
    def complete_task(self):
        print (f"{self.name} has completed a task")
```

```python
jai=manager("Jai")
vk=team_member("vk")
pk=team_member("pk")
tk=team_member("tk")
```

```python
jai.login()
jai.view_tasks()
jai.assign_task()
vk.login()
vk.view_tasks()
vk.complete_task()
```
